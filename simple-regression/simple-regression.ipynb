```python
import pandas as pd
import numpy as np
from ludwig.api import LudwigModel
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'ModelYear', 'Origin']

df = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)
```

```python
df['split'] = np.random.randn(df.shape[0], 1)
msk = np.random.rand(len(df)) <= 0.7

train_df = df[msk]
test_df = df[~msk]

train_df.head()
```

```python
model = LudwigModel(config='/kaggle/input/simpl-reg-yaml/config.yaml')

df.head()
```

```python
print("Training started...")
results = model.train(dataset=train_df)
print("Training completed!")
```

```python
model = LudwigModel.load('/kaggle/working/results/api_experiment_run_0/model')
predictions, _ = model.predict(dataset=test_df)
predictions.head()
```
