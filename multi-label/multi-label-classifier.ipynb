```python
from ludwig.api import LudwigModel
import pandas

df = pandas.read_csv('/kaggle/input/multilabel/multilabel_classification.csv')
model = LudwigModel(config='/kaggle/input/multilabel/multi-label.yaml')

df.head()
```

```python
# To add the file path to images
df['Image_Name'] = '/kaggle/input/multilabel/images/images/' + df['Image_Name'].astype(str)
df.columns = ["image_path", "tags"]
df.head()
```

```python
results = model.train(dataset=df)
```
